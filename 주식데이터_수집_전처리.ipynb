### 시가총액 데이터셋 marcap

https://github.com/FinanceData/marcap

컬럼 구성(18개)

Date : 날짜 (DatetimeIndex)  
Rank: 시가총액 순위 (당일)  
Code : 종목코드  
Name : 종명이름  
Open : 시가  
High : 고가  
Low : 저가  
Close : 종가  
Volume : 거래량  
Amount : 거래대금  
Changes : 전일대비  
ChangeCode: 등락 기호  
ChagesRatio : 전일대비 등락률  
Marcap : 시가총액(백만원)  
Stocks : 상장주식수  
MarketId : 시장기호  
Market : 시장  
Dept : 부서(한국거래소)  

```python
import pandas as pd

mcp = pd.read_csv('./raw/marcap-2019.csv')
mcp
```

```python
mcp.info()
```

```python
mcp.drop(['Dept','MarketId','Amount','Stocks'], axis=1)
```

```python
# 19년 1월 2일 기준 KOSPI 시가총액 상위 100개
mcp1 = mcp[(mcp['Date'] == '2019-01-02') & (mcp['Marcap'][(mcp['Date']=='2019-01-02')&(mcp['Market']=='KOSPI')].rank(ascending=False) <= 100)]
```

```python
mcp1[['Code', 'Name']]
```

```python
codes = mcp1['Code'].values
```

```python
df = mcp[mcp['Code'].isin(list(codes))]
```

```python
df = df.reset_index().drop(['index'], axis=1)
df = df.drop(['Dept'], axis=1)
```

```python
df_stock = df.copy()
```

```python
df_stock
```

ChangeCode : 기본 라벨값  
1: 전일대비 상승  1 ->  1 유지
2 : 전일대비 하락 2  -> 0으로 변경
3 : 등락 변동 없음 3 -> 1로 변경 (손실이 없었다라고 판단)

```python
df_stock['ChangeCode'][df_stock['ChangeCode']==2] = 0
df_stock['ChangeCode'][df_stock['ChangeCode']==3] = 1
```

```python
df_stock1 = df_stock[['Code','Name','ChangeCode','Date']]
df_stock1 = df_stock1.sort_values(by=['Code','Date'])
df_stock1.to_csv('./전처리/df_stock.csv', encoding='utf-8')
```

+0.5% 부터 +30%  : 1  증가  -> 비중 34%
